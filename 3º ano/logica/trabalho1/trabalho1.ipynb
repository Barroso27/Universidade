## Explicação do problema
Este projeto tem como objetivo construir um horario semanal para planear reuniões de projeto de uma "StartUp", seguindo as seguintes condições:
 * Cada reunião só pode ocupar uma sala durante um determinado tempo e dia
 * Cada reunião esta associada a um projeto
 * Cada Projeto tem um conjunto de colaboradores
 * Um dos colaboradores é o líder
 * Cada projeto tem  um numero de reuniões semanais
 * O lider tem de participar em todas as reuniões do seu projeto
 * Os colaboradores podem ou nao participar nas reuniões
 * Em cada reunião tem de ter pelo menos 50% dos colaoradores
 


# Variaveis
#### Inputs
- S - Sala
- D - Dia
- T - Tempo
- P - Projeto
- C - colaboradores
- $ Constprojeto_{c,l,r} $ - Representa os colaboradores C associados a cada projeto, com um líder L designado e um número de reuniões semanais R.
- $ Dispcolab_{c,d,t} $ - Representa a disponibilidade do colaborador C no dia D durante o tempo T.

#### Auxiliares

- $ sl_{d,t,p,s} $ - Representa a atribuição de uma sala S a um projeto P, que ocorre no dia D e no tempo T
- $ slc_{c,d,t,p,s} $ - representa a alocação de um dado colaborador C, num projeto P, a decorrer no dia D, no tempo T e na sala S

#### Matriz de alocação de reuniões

A matriz sl serve para alocar reuniões no dia D no tempo T, do projeto P na sala S, logo temos
$$
\forall {\scriptstyle d \in D}, \, \forall {\scriptstyle t \in T}, \, \forall {\scriptstyle p \in P}, \, \forall {\scriptstyle s \in S} \, \text{sl}_{d,t,p,s} = 1
$$
se existir no dia D e no tempo T uma reunião P na sala S

##### Matriz de alocação de colaboradores a uma reunião

A matriz slc serve para alocar colaboradores a reuniões, logo temos
$$
\forall {\scriptstyle c \in C}, \, \forall {\scriptstyle d \in D}, \, \forall {\scriptstyle t \in T}, \, \forall {\scriptstyle p \in P}, \, \forall {\scriptstyle s \in S}, \, \text{slc}_{c,d,t,p,s} = 1
$$
se um colaborador tem no dia D, tempo T, uma reunião P na sala S 

## Restrições

1. Se o colaborador não tiver disponibilidade, não lhe é atribuída essa opção 

$$
\forall {\scriptstyle p \in P} \, \forall {\scriptstyle c \in \text{Constprojeto}[p][0]} \, \forall {\scriptstyle d \in D} \, \forall {\scriptstyle t \in T} \, \forall {\scriptstyle s \in S}, \, (t,d) \notin \text{Dispcolab}_{c} \Rightarrow \text{slc}_{c,d,t,p,s} = 0
$$

2. Cada projeto tem de ter o seu lider presente em todas as reuniões

$$
\forall {\scriptstyle p \in P}, \, \forall {\scriptstyle d \in D}, \, \forall {\scriptstyle t \in T}, \, \forall {\scriptstyle s \in S}, \, 
(t, d) \notin \text{Dispcolab}[\text{Constprojeto}[p][1]] \Rightarrow \text{sl}_{d,t,p,s} = 0
$$

3. Cada projeto tem um certo número de reuniões semanais

$$
\forall {\scriptstyle p \in P}, \quad \sum_{d \in D, \, t \in T, \, s \in S} \text{sl}_{d,t,p,s} = \text{Constprojeto}[p][2]

$$

4. Cada reunião tem de ter, no minimo, 50% de assiduidade

$$
\forall {\scriptstyle p \in P}, \, \forall {\scriptstyle d \in D}, \, \forall {\scriptstyle t \in T}, \, \forall {\scriptstyle s \in S}, \quad
\sum_{c \in \text{Colaboradores do Projeto } p} slc[c, d, t, p, s] \geq \max\left( \left\lceil \frac{\text{len(Constprojeto[p][0])}}{2} \right\rceil, 2 \right) \cdot sl[d, t, p, s]
$$

5. Cada sala só pode ter uma reunião num certo dia e horário

$$
\forall {\scriptstyle d \in D}, \, \forall {\scriptstyle t \in T}, \, \forall {\scriptstyle s \in S}\quad
\sum_{p \in P} sl[d, t, p, s] \leq 1
$$

6. Cada colaborador só pode participar numa reunião por horário 

$$
\forall {\scriptstyle c \in C}, \, \forall {\scriptstyle d \in D}, \, \forall {\scriptstyle t \in T} \quad
\sum_{ p \in P} \sum_{s \in S}\quad sl[d, t, p, s] \cdot \text{se}\space(c \in \text{Constprojeto}[p][0]) \leq 1
$$


```python
from ortools.linear_solver import pywraplp
import math
from tabulate import tabulate

def h(S, P, D, T, C, Constprojeto, Dispcolab):
    
    solver = pywraplp.Solver.CreateSolver('SCIP') # Bom para problemas de otimização inteira mista

    if not solver:
        print("Solver SCIP não disponível.")
        return
    #SCIP-Programação Linear: Problemas onde as variáveis podem assumir valores contínuos.
    #-Programação Inteira Mista (MIP): Problemas onde algumas variáveis devem ser inteiras e outras podem ser contínuas.

    #Criação de variáveis: O solver permite que você crie variáveis, sejam elas contínuas, inteiras ou booleanas.
    #Solução: Uma vez que as variáveis, restrições e objetivo estão definidos, o solver busca uma solução que satisfaça todas as restrições e que otimize o valor da função objetivo.
    
    # Criar variáveis de reuniões por dia, horário, projeto e sala
    sl = {} 
    for d in D:
        for t in T:
            for p in P:
                for s in S:
                    sl[d, t, p, s] = solver.BoolVar(f'sl[{d}][{t}][{p}][{s}]')
    
    #Para cada combinação de d, t, p e s, uma variável booleana é criada no solver usando a função solver.BoolVar(). O argumento fornecido para essa função é uma string formatada f'sl[{d}][{t}][{p}][{s}]', que serve como o nome dessa variável.
    #Essa variável booleana sl[d, t, p, s] pode assumir os valores 0 ou 1, e o solver vai determinar o valor apropriado conforme a solução do problema de otimização.
    # Criar variáveis de disponibilidade de colaboradores para as reuniões
    slc = {}  
    for p in P:
        for c in Constprojeto[p][0]:
            for d in D:
                for t in T:
                    for s in S:
                        slc[c, d, t, p, s] = solver.BoolVar(f'slc[{c}][{d}][{t}][{p}][{s}]')

    # 1. Adicionar restrição de disponibilidade de colaboradores
    for p in P:
        for c in Constprojeto[p][0]:
            for d in D:
                for t in T:
                    for s in S:
                        if (t, d) not in Dispcolab[c]:
                            solver.Add(slc[c, d, t, p, s] == 0)
                            #se aquela certa hora e dia nao é disponivel para c entao Esta linha adiciona uma restrição ao solver, forçando que a variável de decisão slc[c, d, t, p, s] seja igual a 0.


    # 2. Adicionar restrição de que o líder do projeto deve estar presente
    for p in P:
        for d in D:
            for t in T:
                for s in S:
                    lider = Constprojeto[p][1]
                    if (t, d) not in Dispcolab[lider]:
                        solver.Add(sl[d, t, p, s] == 0)
    #solver.Add(sl[d, t, p, s] == 0): Esta linha adiciona uma restrição ao solver, forçando que a variável de decisão sl[d, t, p, s] seja igual a 0.
#sl[d, t, p, s] provavelmente representa uma variável binária que indica se o colaborador s está alocado no projeto p para a demanda d no tempo t.
#Se o líder não estiver disponível para essa combinação de tempo t e demanda d, então a variável é forçada a 0, proibindo essa alocação.


    # 3. Adicionar restrição onde existem as reuniões semanais necessárias para cada projeto
    for p in P:
        solver.Add(
            solver.Sum(sl[d, t, p, s] for d in D for t in T for s in S) == Constprojeto[p][2]
        )
    #sumamos todas as alocações feitas no projeto p (considerando cada demanda, tempo e colaborador),Essa soma deve ser igual ao valor Constprojeto[p][2]
    # 4. Ajustar restrição onde pelo menos 50% dos colaboradores e mínimo de 2 devem estar presentes
    for p in P:
        for d in D:
            for t in T:
                for s in S:
                    necessarios = math.ceil(len(Constprojeto[p][0]) / 2)
                    solver.Add(
                        solver.Sum(slc[c, d, t, p, s] for c in Constprojeto[p][0]) >= max(necessarios,2) * sl[d, t, p, s]
                    )
    #math.ceil(len(Constprojeto[p][0]) / 2) calcula a metade desse número e arredonda para cima.
    #solver.Sum(slc[c, d, t, p, s] for c in Constprojeto[p][0])Essa expressão soma todas as alocações slc[c, d, t, p, s] dos colaboradores possíveis (definidos em Constprojeto[p][0]) 
    #max(necessarios, 2) retorna o maior valor entre necessarios e 2. Assim, garante-se que sempre haja pelo menos 2 alocações.
    # se o sl for igual 1 quer dizer que pode ser nesse horario e que a soma de colaboradores slc deve ser pelo menos necessarios ou 2.
    #caso seja 0 Se sl[d, t, p, s] == 0, isso significa que o colaborador principal não está alocado para a demanda d, no tempo t, para o projeto 
    #A multiplicação por sl[d, t, p, s] é o que "liga" ou "desliga" a necessidade de alocação de colaboradores. Se sl[d, t, p, s] == 0, o lado direito da restrição (o mínimo de colaboradores exigido) é multiplicado por zero, o que resulta em zero, ou seja, nenhum colaborador será necessário ou alocado.
    
    # 5. Evitar conflitos de sala: apenas uma reunião por sala, dia e horário
    for d in D:
        for t in T:
            for s in S:
                solver.Add(
                    solver.Sum(sl[d, t, p, s] for p in P) <= 1
                )

    # 6. Garantir que cada colaborador participe de no máximo uma reunião por horário
    for c in C:
        for d in D:
            for t in T:
                solver.Add(
                    solver.Sum(sl[d, t, p, s] for p in P for s in S if c in Constprojeto[p][0]) <= 1
                )
                
    #<= 1:Esta é a restrição que assegura que a soma das variáveis sl[d, t, p, s] (ou seja, o número de projetos alocados ao colaborador s para a demanda d no tempo t) não pode ser maior que 1. Isso significa que o colaborador pode, no máximo, estar alocado em um único projeto ao mesmo tempo.
    #D = {Demanda1}
    #T = {Tempo1, Tempo2}
    #P = {Projeto1, Projeto2}
    #S = {Colaborador1, Colaborador2}
#Agora, ao aplicar a restrição para o Colaborador1, a soma das variáveis sl[Demanda1, Tempo1, Projeto1, Colaborador1] e sl[Demanda1, Tempo1, Projeto2, Colaborador1] deve ser no máximo 1. Ou seja, o Colaborador1 pode estar alocado ou no Projeto1 ou no Projeto2, mas não em ambos ao mesmo tempo.
    status = solver.Solve()
    #chamada ao Método Solve:

    #solver.Solve() chama o método que inicia o processo de resolução do problema que foi definido através de variáveis, restrições e funções objetivas.
    #Processamento do Problema:O solver analisa as variáveis e restrições que você definiu anteriormente no seu código. Ele utiliza algoritmos de otimização para encontrar uma solução que maximize ou minimize a função objetivo (dependendo de como você a definiu) enquanto atende a todas as restrições.
    #Armazenamento do Status:O resultado da chamada ao método Solve() é armazenado na variável status. O valor dessa variável indicará o resultado do processo de resolução.

    if status == pywraplp.Solver.OPTIMAL:
        cabeça = ['Dia', 'Hora', 'Projeto', 'Sala', 'Participantes']
        tabela = []
        for d in D:
            for t in T:
                for p in P:
                    for s in S:
                        if sl[d, t, p, s].solution_value() > 0:
                            #Para cada combinação, a condição verifica se a variável sl[d, t, p, s] tem um valor maior que zero. Isso indica que a alocação foi feita (ou seja, o colaborador está alocado na sala s no dia d e hora t para o projeto p).
                            participantes = [c for c in Constprojeto[p][0] if slc[c, d, t, p, s].solution_value() > 0]
                            #Se a alocação for positiva, o código então coleta a lista de colaboradores (participantes) que foram alocados para o projeto p no dia d e hora t na sala s. Isso é feito verificando a variável slc, que provavelmente representa a alocação de colaboradores
                            tabela.append([d, t, p, s, participantes])
        print(tabulate(tabela, headers=cabeça))
```
